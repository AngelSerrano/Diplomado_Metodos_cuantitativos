```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ipc=pd.read_csv('IPC.csv')
ipc.Date =pd.to_datetime(ipc.Date, infer_datetime_format=True)
ipc.set_index('Date',inplace=True)
```

```python
cols=['Open','Close']
plt.plot(ipc['Open'],label = 'Open');
plt.plot(ipc['Close'],label = 'Close',color='r',linestyle='-.',linewidth=2);
plt.legend();
```

```python
f, ax1,  = plt.subplots(figsize=(20, 10))
ax1.plot(ipc.index,ipc['Open'],label = 'Open',color = 'black');
ax1.plot(ipc.index,ipc['Close'],label = 'Close',color='r',linestyle='-.',linewidth=2);
plt.legend(loc=0);
ax2 =ax1.twinx()
ax2.bar(ipc.index,ipc['Volume']/1000,label = 'volume')
plt.legend(loc =2);
f.autofmt_xdate()
plt.show()
```
